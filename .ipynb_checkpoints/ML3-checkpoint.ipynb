```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
mapa = {
    "unfinished" : "nao_finalizado",
    "expected_hours" : "horas_estipuladas",
    "price" : "preco"
}
dados = dados.rename(columns = mapa)
dados.head()
```

```python
trocar = {
    0 : 1,
    1 : 0
}
dados['finalizado'] = dados.nao_finalizado.map(trocar)
dados.tail()
```

```python
import seaborn as sns

sns.scatterplot(x='horas_estipuladas', y='preco', data=dados)
```

```python
sns.scatterplot(x='horas_estipuladas', y='preco', hue='finalizado', data=dados)
```

```python
sns.relplot(x='horas_estipuladas', y='preco', hue='finalizado', col='finalizado', data=dados)
```

```python
x = dados[['horas_estipuladas', 'preco']]
y = dados ['finalizado']
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

SEED = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.3, stratify = y)
print("O modelo foi treinado com %d elementos e testado com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi de %.2f%%" % acuracia)
```

```python
import numpy as np

previsoes_de_base = np.ones(648)
baseline = accuracy_score(teste_y, previsoes_de_base) * 100
print("A acurácia do algoritmo de baseline foi de %.2f%%" % baseline)
```

```python
sns.scatterplot(x='horas_estipuladas', y='preco', hue=teste_y, data=teste_x)
```

```python
x_min = teste_x.horas_estipuladas.min()
x_max = teste_x.horas_estipuladas.max()
y_min = teste_x.preco.min()
y_max = teste_x.preco.max()
print(x_min, x_max, y_min, y_max)
```

```python
pixels = 100
eixo_x = np.arange(x_min, x_max, (x_max - x_min) / pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min) / pixels)
```

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
```

```python
Z = modelo.predict(pontos)
Z = Z.reshape(xx.shape)
Z
```

```python
import matplotlib.pyplot as plt

plt.contourf(xx,yy,Z,alpha=0.3)
plt.scatter(teste_x.horas_estipuladas, teste_x.preco, c=teste_y, s=1)
```
